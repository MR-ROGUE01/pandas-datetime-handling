```python
def session(x):
    if x < 12:
        return 'Morning'
    elif x < 17:
        return 'Afternoon'
    elif x < 21:
        return 'Evening'
    else:
        return 'Night'

time['session'] = time['hour'].apply(session)
```

```python
import datetime

today = datetime.datetime.today()

today - time['date']
```
